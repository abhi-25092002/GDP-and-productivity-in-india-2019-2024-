# DATA VISULIZATION :-

# 1.Which Year highest and lowest analysis of GDP and productivity :-

```python
sns.scatterplot(df['Year'],color='darkgreen')
```

# 2.Comparison Between both Unemploment Ratio :-

```python
plt.subplots(figsize=(15,9))
sns.swarmplot(x='Unemployment Rate (%)',y='Youth Unemployment Rate (%)',data=df,color='r')
plt.grid()
plt.title('COMPARISON BETWEEN Unemployment Rate (%) AND Youth Unemployment Rate (%)',color='g')
```

# 3.City regarding R&D Expenditure (% of GDP) in india (2019-2024) :-

```python
plt.subplots(figsize=(15,8))
ax=sns.barplot(y=df['R&D Expenditure (% of GDP)'],x=df['City'])
ax.set_xticklabels(ax.get_xticklabels(),rotation=35)
plt.xlabel('R&D Expenditure (% of GDP)')
plt.ylabel('City')
plt.title('City regarding R&D Expenditure (% of GDP) in india (2019-2024)',color='r')
```

# 4.Analysis about density of Patents per 100,000 Inhabitants :-

```python
sns.distplot(df['Patents per 100,000 Inhabitants'],color='brown')
```

```python
df.head()
```

# 5.Comparison between Unemployment Rate (%) and ICT Sector Employment (%) :-

```python
sns.jointplot(x='Unemployment Rate (%)',y='SME Employment (%)',data=df,color='darkred')
plt.grid(color='black')
```

# 6.Which number of R&D Expenditure (% of GDP) highest and lowest :-

```python
df.groupby('R&D Expenditure (% of GDP)')['City'].count().nlargest(16).plot(kind = 'pie', autopct = '%1.1f%%')
```

# 7.Analysis of SME employment(%) according to Cities :-

```python
plt.subplots(figsize=(30,15))
sns.violinplot(x=df['City'],y=df['SME Employment (%)'])
plt.xlabel('City',size=20,color='darkgreen')
plt.ylabel('SME Employment (%)',size=20,color='darkgreen')
```

# 8.Analysis of City regarding ICT Sector Employment (%) :-

```python
plt.subplots(figsize=(18,10))
ax=sns.boxplot(x='City', y='ICT Sector Employment (%)', data=df,palette='rainbow')
ax.set_xticklabels(ax.get_xticklabels(),rotation=55)
```

# 9.All Employment ratio comparison with year :-

```python
df2=df[['SME Employment (%)','Tourism Sector Employment (%)','ICT Sector Employment (%)','Year']]
df2.head(2)
```

```python
sns.pairplot(df2,hue='Year',height=4)
```

# 10.Both Unemployment ratio analysis regarding Cities :-

```python
df3=df[['Unemployment Rate (%)','Youth Unemployment Rate (%)','City']]
df3.head()
```

```python
sns.pairplot(df3,hue='City',height=2.5)
```
